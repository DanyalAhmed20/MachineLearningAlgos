```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
df['State']=le.fit_transform(df['State'])
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1:].values
```

```python
x
```

```python
ohe = OneHotEncoder()
z=ohe.fit_transform(x[:,3:4]).toarray()
x=np.delete(x,3,axis=1)
z=np.delete(z,0,axis=1)
x=np.concatenate((z,x),axis=1)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
xtrain
```

```python
xtrain.shape
```

```python
import keras
from keras.models import Sequential #initializing the neural network
from keras.layers import Dense #in creating the nn layers
```

```python
reg = Sequential()
```

```python
reg.add(Dense(units=5,init="random_uniform",activation='relu'))#1st layer
```

```python
reg.add(Dense(units=12,init="random_uniform",activation='relu'))#2nd layer
```

```python
reg.add(Dense(units=8,init="random_uniform",activation='relu')) #3rd layer
```

```python
reg.add(Dense(units=1,init="random_uniform",activation='relu')) #output
```

```python
reg.compile(optimizer='adam', loss='mse', metrics=['mse'])
```

```python
reg.fit(xtrain,ytrain,batch_size=10,epochs=300)
```

```python
reg.summary()
```

```python
ypredreg = reg.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypredreg)
```

```python
reg.save("50_startreg.h5")
```
