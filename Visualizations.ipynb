# Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(5,20,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

### Line plot

```python
plt.plot(x,y,"g*")
plt.xlabel("Integers")
plt.ylabel("Squares of Integers")
plt.title("Graph between int and squares")
```

```python
plt.scatter(x,y,color="green")
plt.xlabel("Integers")
plt.ylabel("Squares of Integers")
plt.title("Graph between int and squares")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"green")
plt.xlabel("Integers")
plt.ylabel("Squares of Integers")
plt.title("Graph between int and squares")
plt.subplot(1,2,2)
plt.scatter(x,y,color="green")
plt.xlabel("Integers")
plt.ylabel("Squares of Integers")
plt.title("Graph between int and squares")
```

```python
plt.plot(x,y,"green", label = "b/w x and y")
plt.plot(y,x,"red", label = "b/w y and x")
plt.xlabel("Integers")
plt.ylabel("Squares of Integers")
plt.title("Graph between int and squares")
plt.legend(loc="center")
```

```python
plt.bar(x,y,color="green")
plt.xlabel("Integers")
plt.ylabel("Squares of Integers")
plt.title("Graph between int and squares")
```

## Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="day", palette="coolwarm")
```

```python
sns.lineplot(x="size", y="tip", data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.heatmap(tips.corr(), annot=True)
```

```python
tips.corr()
```

```python
sns.barplot(x='sex', y='tip', data=tips)
```

```python
sns.countplot('time', data=tips)
```

```python
sns.boxplot(x="day",y="total_bill", data=tips)
```
